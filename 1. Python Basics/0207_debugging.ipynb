```
import random
```

```
def factorial(x):
  if(x==0):
    return 1
  return x*factorial(x-1)
```

```
factorial(4),factorial(6)
```

```
def code_to_debug():
  for i in range(10):
    x = random.random()
    factorial(x)
```

```
code_to_debug()
```

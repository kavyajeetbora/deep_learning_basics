```
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set()
```

# Plotting Vectors

```
plt.quiver(0,0,3,4)
plt.show()
```

```
plt.quiver(0,0,6,4,scale_units='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,1,4,scale_units='xy',angles='xy',scale=1,color='b')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```
def plot_vectors(vectors):
  i=0
  colors = ['r','b','g','y']
  for vector in vectors:
    plt.quiver(vector[0],vector[1],vector[2],vector[3],scale_units='xy',angles='xy',scale=1,color=colors[i%len(colors)])
    i += 1
  plt.xlim(-10,10)
  plt.ylim(-10,10)
  plt.show()
```

```
vectors = [(0,0,4,5),(0,0,3,2),(0,0,-5,2),(0,0,-5,-4)]
plot_vectors(vectors)
```

# Vector addition and subtraction

```
vectors = [np.asarray([0,0,4,5]),np.asarray([0,0,3,2])]
plot_vectors(vectors)
```

```
vec_sum = vectors[0] + vectors[1]
vec_sum
```

```
vectors.append(vec_sum)
plot_vectors(vectors)
```

```
plot_vectors([vectors[0], vectors[1], vectors[0]-vectors[1]])
```

# Vector dot product

```
a1 = np.asarray([0,0,3,4]) # red 
b1 = np.asarray([0,0,4,-1]) # blue

vectors = [a1,b1]
plot_vectors(vectors)
```

```
a = np.asarray([3,4])
b = np.asarray([4,-1])
np.dot(a,b)
```

```
# projection of the vector a on b
a_b = np.dot(a,b)/np.linalg.norm(b)
a_b
```

```
b_unit = b/np.linalg.norm(b)
b_unit
```

```
a_proj = a_b*b_unit
a_proj
```

```
#projection of a vector on b
a2 = np.asarray([0,0,a_proj[0],a_proj[1]])  # green
vectors = [a1,b1,a2]
plot_vectors(vectors)
```

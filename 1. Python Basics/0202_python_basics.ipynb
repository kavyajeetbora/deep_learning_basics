# List

Python map() function is used to apply a function on all the elements of specified iterable and return map object. Python map object is an iterator, so we can iterate over its elements. We can also convert map object to sequence objects such as list, tuple etc. using their factory functions

```
names = ['kavyajeet','bora','kabeer','kristy']
def toUpper(s):
  return s.upper()
```

```
b = list(map(toUpper,names))
b
```

```
# we can also use lambda functions
c = list(map(lambda x: x.lower(),b))
c
```

```
# another function in list is we can filter the items
c = list(filter(lambda x: x.startswith('k'),c))
c
```

# Tuples

```
# they are like list but they cannot be assigned after defining 
import time
mylist = list(range(100000))
tic = time.time()
my_list_squared = list(map(lambda x: x**2,mylist))
toc = time.time()
print('time taken to square the list: ',toc-tic)
```

```
# they are like list but they cannot be assigned after defining 
import time
mytuple = tuple(range(100000))
tic = time.time()
my_tuple_squared = list(map(lambda x: x**2,mytuple))
toc = time.time()
print('time taken to square the list: ',toc-tic)
```

operations on tuples are faster than in list

# Sets

```
my_phone = ['Iphone5s',False,4.3,'3G','single_sim',35000]

myset = set(my_phone)
myset
```

```
my_phone.append(4.3)
my_2nd_set = set(my_phone)
print(my_phone,'\n',my_2nd_set)
```

the sets contains only the unique items in them

```
# lets see the time taken in operations done in the set and compare it with lists
my_list1 = list(range(1000000))
tic = time.time()
print(123124 in my_list1)
print(12414 in my_list1)
print(111111111 in my_list1)
toc = time.time()
print('Time elapsed: ',toc-tic)
```

```
my_set1 = set(range(1000000))
tic = time.time()
print(123124 in my_set1)
print(12414 in my_set1)
print(111111111 in my_set1)
toc = time.time()
print('Time elapsed: ',toc-tic)
```

we search and see if the unique item is there in the collection using sets very very quickly

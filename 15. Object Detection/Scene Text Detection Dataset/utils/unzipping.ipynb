```
def extract_files(zip_file, destination):
  if os.path.isdir(destination):
    shutil.rmtree(destination)
  os.mkdir(destination)

  with ZipFile(zip_file,'r') as file:
    file.extractall(destination)
  os.listdir(destination)
  print('Unzipped ',len(os.listdir(destination)), 'successfully')

img_destination = '/content/drive/My Drive/Colab Notebooks (1)/padh.ai.notebooks/15. Object Detection/AI4Bharat Dataset/Training Set/Images'
annotation_destination = '/content/drive/My Drive/Colab Notebooks (1)/padh.ai.notebooks/15. Object Detection/AI4Bharat Dataset/Training Set/Annotatons'
extract_files(image_zip_file,img_destination)
```
